```python
class Bank:
  def __init__(self,balance):
    self.balance=balance

  def deposit(self, amount):
      self.balance=self.balance+amount

  def get_balance(self):
      return self.balance

acc=Bank(1000)
acc.deposit(500)
print("Balance:",acc.get_balance())
```

```python
class Calculator:
  def add(self, a,b):
      print("parent class: Adding numbers")
      return a+b

class advancedCalculator(Calculator):
    def add(self, a,b ):
      result = super().add(a,b)
      print("child class: adding 10 extra")
      return result+10

calc = advancedCalculator()
print("final result:", calc.add(5,3))
```

```python
class Dog:
  def sound(self):
      return "Bark"
class Cat:
  def sound(self):
      return "Meow"
class Cow:
  def sound(self):
      return "Moo"

animals = [Dog(), Cat(), Cow()]

for animal in animals:
  print(animal.sound())
```
